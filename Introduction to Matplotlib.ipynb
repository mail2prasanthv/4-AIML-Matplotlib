 # Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare Data
x = [1,2,3,4]
y = [20,21,23,24]


#2. setup plot
fig, ax = plt.subplots(figsize=(1,1))#width, height

#3. plot data
ax.plot(x, y)

#4. customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

#5. 
fig.savefig("simple-plot.png")
```
