 # Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare Data
x = [1,2,3,4]
y = [20,21,23,24]


#2. setup plot
fig, ax = plt.subplots(figsize=(1,1))#width, height

#3. plot data
ax.plot(x, y)

#4. customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

#5. 
fig.savefig("simple-plot.png")
```

## Making figures with Numpy Arrays

We want : 
    * Line Plot
    * Scatter Plot
    * Bar Plot
    * Histogram
    * Subplots
    

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100) # returns 100 evenly spaced numbers between 0 and 10
x
```

## Line Plot

```python
fig, ax = plt.subplots(figsize=(4,4))#width, height
ax.plot(x, x**2);# x as x cordinate and x^2 as y cordinate
```

## Scatter Plot

```python
fig, ax = plt.subplots(figsize=(7,3))#width, height
ax.scatter(x, np.exp(x))# x as x cordinate and x^2 as y cordinate
```

```python
fig, ax = plt.subplots(figsize=(7,3))#width, height
ax.scatter(x, np.sin(x))# x as x cordinate and x^2 as y cordinate
```

# Bar plot

```python
price_data = {"Almond Butter": 10,
               "Peanut Butter": 20,
              "Cashew Butter": 30} # python dictionary

fig, ax = plt.subplots() #width, height
ax.bar( price_data.keys(), price_data.values())# x as x cordinate and height as y coordinate 
ax.set(title="Butter Price", ylabel="Price $");
```

# Histogram

```python
x = np.random.randn(1000)#Return a sample (or samples) from the standard normal distribution.
fig, ax = plt.subplots()
ax.hist(x)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols= 2, figsize=(10,5))

x = np.random.randn(1000)#Return a sample (or samples) from the standard normal distribution.
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar( price_data.keys(), price_data.values())
ax4.hist(x)
```
